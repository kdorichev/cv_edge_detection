# Edge Detection using OpenCV library

```python
import cv2
from matplotlib import pyplot as plt
```

#### Read the source image as grayscale

```python
img = cv2.imread('cv_edge_detection/lena.png',0)
```

```python
plt.figure(figsize=(7,7))
plt.axis('Off')
plt.imshow(img, cmap='Greys_r');
```

```python
# Altervatively use OpenCV for image display
# cv2.namedWindow('image', cv2.WINDOW_NORMAL)
# cv2.imshow('image',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

#### Use Canny function for edge detection

```python
imgEdge = cv2.Canny(img,80,200)
plt.figure(figsize=(7,7))
plt.axis('Off')
plt.imshow(imgEdge, cmap='Greys_r');
```

#### Display inverted

```python
plt.figure(figsize=(7,7))
plt.axis('Off')
plt.imshow(imgEdge, cmap='Greys');
```
